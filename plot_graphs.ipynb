```python
import matplotlib.pyplot as plt
import json

# Load the data from the new file
file_path = '/Users/tannicholas/SLT_FYP/Gloss-Free TiTok freezed mbart free/log.txt'
data = []

# Read and parse the log data
with open(file_path, 'r') as file:
    for line in file:
        data.append(json.loads(line))

# Extract relevant fields for plotting
epochs = [entry['epoch'] for entry in data]
train_loss = [entry['train_loss'] for entry in data]
test_loss = [entry['test_loss'] for entry in data]

# Plot train and test loss curves
plt.figure(figsize=(10, 6))
plt.plot(epochs, train_loss, label='Train Loss', marker='')
plt.plot(epochs, test_loss, label='Test Loss', marker='')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Loss Curves for frozen TiTok and fine-tuned mBART')
plt.legend()
plt.grid(True)
plt.tight_layout()

# Display the plot

# Save the plot as a PDF file
plt.savefig('loss_TiTok_frozen_MBart_free.pdf', format='pdf')
plt.show()
```
